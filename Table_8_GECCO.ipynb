```python
import os
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
model, nfeat, maf, censoring, cv, date = 'het', 'f100', 'maf0.4', '0.1', '0', '2024-04-07'
if model == 'add':
    model_long = 'additive'
elif model == 'epi':
    model_long = '2way_epistasis'
elif model == 'het':
    model_long = 'heterogeneous'
elif model == 'me':
    model_long = 'main_effect'
```

```python
objects = []
with (open('pipeline/pickled_cv_models/'+model+'/'+model+'_'+nfeat+'_'+maf+'/cens_'+censoring+'/ExSTraCS_'+cv+'_'+date, "rb")) as file:
    while True:
        try:
            objects.append(pickle.load(file))
        except EOFError:
            break
```

```python
trainedModel = objects[0]
```

```python
#load data
data = pd.read_csv('pipeline/cv_sim_data/cv_'+model+'/'+model+'_'+nfeat+'_'+maf+\
    '/'+model_long+'_cens'+censoring+'_surv_'+date+'_CV_'+cv+'_Train.txt', sep = '\t', header = 0)
timeLabel = 'eventTime'
censorLabel = 'eventStatus'
inst = 'inst'

dataHeaders = data.drop([timeLabel,censorLabel, inst],axis=1).columns.values
headers = dataHeaders
```

```python
trainedModel.export_final_rule_population(headers,censorLabel,filename="output/popData.csv")
populationData = pd.read_csv("output/popData.csv")
display(populationData.head())
```

```python
populationData = populationData.sort_values('Numerosity', ascending=False)
populationData = populationData
display(populationData)
```

```python
populationData.to_csv('output/Table8.csv')
```
