```python
import os
import numpy as np
import pandas as pd
```

```python
folder_path = '../pipeline/sim_lcs_output'

# List to hold individual DataFrames
data_frames = []

# Traverse the directory structure
for root, _, files in os.walk(folder_path):
    for file in files:
        if 'Perm' not in file and file.endswith('.csv') and 'Cox' in file:
            file_path = os.path.join(root, file)
            # Read CSV into a DataFrame
            df = pd.read_csv(file_path, index_col='times')
            # Append to the list
            data_frames.append(df)
    
    # Concatenate all DataFrames
merged_df = pd.concat(data_frames, axis=1, join='outer')
merged_df
```

```python
# Extract the base labels by removing the `_cvX` part
base_labels = list(merged_df.columns.str.extract(r'(.+)_cv\d+$')[0])
```

```python
# Group by the base label and calculate the mean across columns in each group
mean_df = merged_df.groupby(base_labels, axis=1).mean().reset_index()
mean_df
```

```python
std_df = merged_df.groupby(base_labels, axis=1).std().reset_index()
lower_ci_df = mean_df - 1.96 * std_df
upper_ci_df = mean_df + 1.96 * std_df
```

```python
def fun(df, col_name='b_scores_add_f10000_maf0.2_cens0.1'):
    temp_df = df[[col_name, 'times']]
    temp_df = temp_df.dropna()
    try:
        val = np.trapz(temp_df[col_name], temp_df['times']) / (list(temp_df['times'])[-1] - list(temp_df['times'])[0])
    except Exception as e:
#         print(col_name, e)
        val = np.nan
    return val
```

```python
mean_ibs_dict = dict()
for col in mean_df.columns:
    mean_ibs_dict[col] = [fun(mean_df, col),]
mean_ibs_df = pd.DataFrame(mean_ibs_dict).drop('times', axis=1).T
mean_ibs_df
```

```python
lci_ibs_dict = dict()
for col in lower_ci_df.columns:
    lci_ibs_dict[col] = [fun(lower_ci_df, col),]
lci_ibs_df = pd.DataFrame(lci_ibs_dict).drop('times', axis=1).T
```

```python
uci_ibs_dict = dict()
for col in upper_ci_df.columns:
    uci_ibs_dict[col] = [fun(upper_ci_df, col),]
uci_ibs_df = pd.DataFrame(uci_ibs_dict).drop('times', axis=1).T
```

```python
combined_ibs_df = pd.concat([mean_ibs_df, lci_ibs_df, uci_ibs_df], axis=1)
combined_ibs_df
```

```python
split_labels = combined_ibs_df.index.str[9:].str.split('_', expand=True)
# split_labels.to_frame()
```

```python
final_ibs_df = combined_ibs_df
final_ibs_df.index = split_labels
final_ibs_df.sort_index(level=1, ascending=[True], inplace=True)
```
